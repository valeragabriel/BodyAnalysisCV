```python
# Verifica se a imagem foi carregada com sucesso
if img1 is not None:
    # Setup da instância mediapipe
    mp_pose = mp.solutions.pose
    mp_drawing = mp.solutions.drawing_utils

    with mp_pose.Pose(min_detection_confidence=0.5, min_tracking_confidence=0.5) as pose:
        # Recolor image para RGB
        image = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
        image.flags.writeable = False

        # Faz a detecção
        results = pose.process(image)

        # Recolor de volta para BGR
        image.flags.writeable = True
        image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)

        # Extrai landmarks
        try:
            landmarks = results.pose_landmarks.landmark
            print(landmarks)
        except:
            pass

        # Renderiza os landmarks
        mp_drawing.draw_landmarks(image, results.pose_landmarks, mp_pose.POSE_CONNECTIONS,
                                  mp_drawing.DrawingSpec(color=(245, 117, 66), thickness=2, circle_radius=2),
                                  mp_drawing.DrawingSpec(color=(245, 66, 230), thickness=2, circle_radius=2)
                                  )

        # Exibe a imagem com landmarks
        cv2.imshow('Pose Detection', image)

        # Aguarda pela tecla 'q' por 10 milissegundos
        key = cv2.waitKey(10) & 0xFF

        # Se a tecla 'q' for pressionada, encerra o loop
        if key == ord('q'):
            pass

    cv2.destroyAllWindows()

else:
    print('Erro ao carregar a imagem.')
```

```python
altura, largura, _ = img1.shape

# Definir um ponto no terço superior da imagem
ponto_x = largura // 2
ponto_y = altura // 3

# Desenhar um círculo no ponto definido
cv2.circle(img1, (ponto_x, ponto_y), 5, (0, 255, 0), -1)  # 5 é o raio do círculo, (0, 255, 0) é a cor (verde), -1 preenche o círculo

# Exibir a imagem com o ponto desenhado
cv2.imshow("Imagem com Ponto", img1)
key = cv2.waitKey(10) & 0xFF


if key == ord('q'):
    pass

cv2.destroyAllWindows()
```
